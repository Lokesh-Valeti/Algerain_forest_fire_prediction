```python
x=df_copy.drop("FWI",axis=1)
y=df_copy["FWI"]
```

```python
#Train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x_train.corr()
```

```python
#checking for multi colinearity
plt.figure(figsize=(12,10))
sns.heatmap(x_train.corr(),annot=True)
```

```python
def correlation(dataset,thresold):
  col_corr=set()
  corr_matrix=dataset.corr()
  for i in range(len(corr_matrix.columns)):
    for j in range(i):
      if abs(corr_matrix.iloc[i,j]>thresold):
        colname=corr_matrix.columns[i]
        col_corr.add(colname)
  return col_corr
```

```python
corr_features=correlation(x_train,0.85)
```

```python
corr_features
```

```python
x_train.drop(corr_features,axis=1,inplace=True)
x_test.drop(corr_features,axis=1,inplace=True)
x_train.shape,x_test.shape
```

```python
#@title Feature scaling
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
x_train_scaled
```

```python
#@title Box plot to understand the effect of Standard scaler
```

```python
plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(data=x_train)
plt.title("x_train_before_scaling")
plt.subplot(1,2,2)
sns.boxplot(data=x_train_scaled)
plt.title("x_train after scaling")
```

```python
#@title Linear Regression
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
linregressor=LinearRegression()
linregressor.fit(x_train_scaled,y_train)
y_pred=linregressor.predict(x_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("The mean absloute error =",mae)
print("R2 score =",score)
plt.scatter(y_test,y_pred)
```

```python
#@title Lasso Regression
```

```python
from sklearn.linear_model import Lasso
lasso=Lasso()
lasso.fit(x_train_scaled,y_train)
y_pred=lasso.predict(x_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("The mean absloute error =",mae)
print("R2 score =",score)
plt.scatter(y_test,y_pred)
```

```python
#@title Ridge Regresision moedel
```

```python
from sklearn.linear_model import Ridge
ridge=Ridge()
ridge.fit(x_train_scaled,y_train)
y_pred=ridge.predict(x_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("The mean absloute error =",mae)
print("R2 score =",score)
plt.scatter(y_test,y_pred)
```

```python
from sklearn.linear_model import RidgeCV
ridgecv=RidgeCV(cv=5)
ridgecv.fit(x_train_scaled,y_train)
y_pred=ridgecv.predict(x_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("The mean absloute error =",mae)
print("R2 score =",score)
plt.scatter(y_test,y_pred)
```
